## Renomear as colunas do DataFrame

```python
def rename_columns(dataframe):
    # Cria uma cópia do dataframe para evitar modificações diretas
    renamed_dataframe = dataframe.copy()
    
    # Funções de transformação para as colunas
    title = lambda x: x.title()  # Transforma em formato de título
    snakecase = lambda x: re.sub(r'([a-z])([A-Z])', r'\1_\2', x).lower().replace(" ", "_")  # Snake case manual
    spaces = lambda x: x.replace(" ", "")  # Remove espaços
    
    # Aplica as transformações nas colunas
    cols_old = [title(str(x)) for x in renamed_dataframe.columns]
    cols_old = [spaces(x) for x in cols_old]
    cols_new = [snakecase(x) for x in cols_old]
    
    # Atribui os novos nomes de coluna ao dataframe
    renamed_dataframe.columns = cols_new
    return renamed_dataframe
```

```python
def pais_com_mais(df, coluna_pais, coluna_variavel):
    """
    Encontra o país que possui o maior número de registros em uma variável específica.

    Parâmetros:
    - df (DataFrame): O DataFrame que contém os dados.
    - coluna_pais (str): Nome da coluna que contém os países.
    - coluna_variavel (str): Nome da coluna que contém a variável a ser contada.

    Retorna:
    - Um dicionário com o país e o número máximo de registros.
    """
    # Verificar se as colunas existem no DataFrame
    if coluna_pais not in df.columns or coluna_variavel not in df.columns:
        raise ValueError("As colunas especificadas não estão no DataFrame.")

    # Agrupar por país e contar os registros da variável
    por_pais = df4.groupby(coluna_pais)[coluna_variavel].nunique().reset_index()

    # Identificar o país com mais registros
    maior_pais = por_pais[coluna_variavel].idxmax()
    nome_maior_pais = por_pais.loc[maior_pais, coluna_pais]
    numero = por_pais.loc[:,coluna_variavel].idxmax()
    nome_maior_variavel = por_pais.loc[numero, coluna_variavel]

    # Exibir o resultado
    print(f"O país com mais {coluna_variavel} registrados é {nome_maior_pais}, com {nome_maior_variavel} de {coluna_variavel}.")

    # Retornar os resultados
    return {"pais": nome_maior_pais, "quantidade": nome_maior_variavel}
```

# Visão Cidade

FUNÇÕES CRIADAS PARA ESTÁ VISÃO

```python
def nome_cidade(df, coluna_cidade, coluna_variavel, func):
    """
    Identifica a cidade com o maior número de registros para uma variável específica
    usando uma função agregadora definida.

    Parâmetros:
    - df (DataFrame): O DataFrame que contém os dados.
    - coluna_cidade (str): O nome da coluna que representa as cidades.
    - coluna_variavel (str): O nome da coluna cuja variável será analisada.
    - func (str): O nome da função agregadora a ser usada (e.g., 'count', 'sum').

    Retorna:
    - Um dicionário com o nome da cidade e a quantidade correspondente.
    """

    # Verificar se as colunas existem no DataFrame
    if coluna_cidade not in df.columns or coluna_variavel not in df.columns:
        raise ValueError("As colunas especificadas não estão no DataFrame.")

    # Verificar se a função agregadora é válida
    if not hasattr(pd.core.groupby.DataFrameGroupBy, func):
        raise ValueError("A função agregadora especificada não é válida.")

    # Agrupar por cidade e aplicar a função agregadora
    por_cidade = df.groupby(coluna_cidade).agg({coluna_variavel: func}).reset_index()

    # Identificar a cidade com mais registros
    cidade_idx = por_cidade[coluna_variavel].idxmax()
    nome_cidade = por_cidade.loc[cidade_idx, coluna_cidade]
    maior_quantidade = por_cidade.loc[cidade_idx, coluna_variavel]

    # Exibir o resultado
    print(f"A cidade com mais {coluna_variavel} registrados é {nome_cidade}, com {maior_quantidade} de {coluna_variavel}.")

    # Retornar os resultados
    return {"cidade": nome_cidade, "quantidade": maior_quantidade}
```

```python
def cidades_valores(df, coluna_cidade, coluna_variavel, func, condicao, valor):
    """
    Filtra as cidades com base em uma condição (maior, maior igual, menor ou menor igual) aplicada a uma variável específica
    e calcula um valor agregado (e.g., soma ou contagem).

    Parâmetros:
    - df (DataFrame): O DataFrame que contém os dados.
    - coluna_cidade (str): O nome da coluna que representa as cidades.
    - coluna_variavel (str): O nome da coluna cuja variável será analisada.
    - func (str): O nome da função agregadora a ser usada (e.g., 'count', 'sum').
    - condicao (str): Condição para o filtro ('maior', 'maior igual', 'menor', 'menor igual').
    - valor (int/float): O valor limite para o filtro.

    Retorna:
    - Um DataFrame com as cidades e os valores agregados.
    """

    # Verificar se as colunas existem no DataFrame
    if coluna_cidade not in df.columns or coluna_variavel not in df.columns:
        raise ValueError("As colunas especificadas não estão no DataFrame.")

    # Verificar se a função agregadora é válida
    if not hasattr(pd.core.groupby.DataFrameGroupBy, func):
        raise ValueError("A função agregadora especificada não é válida.")

    # Aplicar a condição para filtrar o DataFrame
    if condicao == "maior":
        df_filtrado = df[df[coluna_variavel] > valor]
    elif condicao == "maior igual":
        df_filtrado = df[df[coluna_variavel] >= valor]
    elif condicao == "menor":
        df_filtrado = df[df[coluna_variavel] < valor]
    elif condicao == "menor igual":
        df_filtrado = df[df[coluna_variavel] <= valor]
    else:
        raise ValueError("A condição deve ser 'maior', 'maior igual', 'menor' ou 'menor igual'.")

    # Agrupar por cidade e aplicar a função agregadora
    resultado = df_filtrado.groupby(coluna_cidade).agg({coluna_variavel: func}).reset_index()

    # Ordenar os resultados do maior para o menor
    resultado = resultado.sort_values(by=coluna_variavel, ascending=False).reset_index(drop=True)

    # Capturar a cidade com o maior valor agregado
    if not resultado.empty:
        cidade_top = resultado.iloc[0][coluna_cidade]
        maior_quantidade = resultado.iloc[0][coluna_variavel]

        # Exibir o resultado
        print(f"A cidade com maior valor de {coluna_variavel} é {cidade_top}, com {maior_quantidade}.")
    else:
        print("Nenhuma cidade atende aos critérios especificados.")
        return None

    # Retornar os resultados
    return resultado
```

```python
def nome_restaurante(df, coluna_restaurante, coluna_variavel, func):
    """
    Identifica o restaurante com o maior número de registros para uma variável específica
    usando uma função agregadora definida.

    Parâmetros:
    - df (DataFrame): O DataFrame que contém os dados.
    - coluna_restaurante (str): O nome da coluna que representa os restaurantes.
    - coluna_variavel (str): O nome da coluna cuja variável será analisada.
    - func (str): O nome da função agregadora a ser usada (e.g., 'count', 'sum').

    Retorna:
    - Um dicionário com o nome do restaurante e a quantidade correspondente.
    """

    # Verificar se as colunas existem no DataFrame
    if coluna_restaurante not in df.columns or coluna_variavel not in df.columns:
        raise ValueError("As colunas especificadas não estão no DataFrame.")

    # Verificar se a função agregadora é válida
    if not hasattr(pd.core.groupby.DataFrameGroupBy, func):
        raise ValueError("A função agregadora especificada não é válida.")

    # Agrupar por cidade e aplicar a função agregadora
    por_restaurante = df.groupby(coluna_restaurante).agg({coluna_variavel: func}).reset_index()

    # Identificar a cidade com mais registros
    restaurante_idx = por_restaurante[coluna_variavel].idxmax()
    nome_restaurante = por_restaurante.loc[restaurante_idx, coluna_restaurante]
    maior_quantidade = por_restaurante.loc[restaurante_idx, coluna_variavel]

    # Exibir o resultado
    print(f"A cidade com mais {coluna_variavel} registrados é {nome_restaurante}, com {maior_quantidade} de {coluna_variavel}.")

    # Retornar os resultados
    return {"cidade": nome_restaurante, "quantidade": maior_quantidade}
```

7. Os restaurantes que fazem reservas são também, na média, os restaurantes que

possuem um valor médio de prato para duas pessoas maior que as churrascarias
